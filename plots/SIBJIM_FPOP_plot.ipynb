```python
from sklearn.metrics import mean_absolute_percentage_error

MAPE_population = mean_absolute_percentage_error(real_population,simulated_data['population'])*100
print("MAPE for population: {} %".format(round(MAPE_population,2)))

MAPE_mortality = mean_absolute_percentage_error(real_mortality,simulated_data['mortality'])*100
print("MAPE for mortality: {} %".format(round(MAPE_mortality,2)))

MAPE_fertility = mean_absolute_percentage_error(real_fertility,simulated_data['fertility'])*100
print("MAPE for fertility: {} %".format(round(MAPE_fertility,2)))

MAPE_emigration = mean_absolute_percentage_error(real_emigration,simulated_data['emigration'])*100
print("MAPE for emigration: {} %".format(round(MAPE_emigration,2)))

MAPE_immigration = mean_absolute_percentage_error(real_immigration,simulated_data['immigration'])*100
print("MAPE for immigration: {} %".format(round(MAPE_immigration,2)))
```

     2022 

```python
real_data_population = pd.read_csv("../data/population_spb_2022", sep=' ')
synth_population = pd.read_csv("../data/people_2022.txt", sep='\t', index_col=0)
age_classes = list(synth_population['age_class'])

age_step = 5
age_classes = 14

def set_class_to_age(row):
    """ Set age class according to age of person"""
    return (row['age']//age_step) if row['age'] <= age_classes*age_step else age_classes
# labels = list(real_data_population['age_group'])
# plt.bar(real_data_population['age_group'], real_data_population['total'])
```

```python
synth_population['age_class'] = synth_population.apply(set_class_to_age, axis=1)
people_age_classes = synth_population.groupby('age_class')['sp_id'].count().to_list()
```

```python
matplotlib.pyplot.rcParams.update({
    # 'text.usetex' : True,
    # 'font.family' : 'STIXGeneral',
    'font.family' : 'sans-serif',
})

plt.rc( 'text', usetex=True ) 
plt.rc('font',family = 'sans-serif',  size=20)

fig, ax = plt.subplots(figsize=(15, 6), tight_layout=True)
width = 1  # the width of the bars
multiplier = 0
x = np.arange(len(people_age_classes)) # the label locations
x = x*3
print(x)
real_data_values = real_data_population['total'].to_list()
synth_pop_values = people_age_classes

values = {
    'Official data': real_data_population['total'].to_list(),
    'Synth. pop.': people_age_classes,   
}

colors = ['orangered', 'royalblue']
percent_labels = [ (round((synth_pop_values[i]/real_data_values[i] - 1)*100, 1)) for i in range(len(synth_pop_values))]
percent_labels = [ r'\textbf{$+' + str(i) + ' \%$}' if i > 0 else '$' + str(i) + ' \%$' for i in percent_labels]

for attribute, measurement in values.items():
    offset = width * multiplier - width/2
    rects = ax.bar(x + offset, measurement, width, label=attribute, color = colors[multiplier])
    if multiplier == 1:
        ax.bar_label(rects, percent_labels, label_type='edge', fontsize=14)
    multiplier += 1

# ax.bar(real_data_population['age_group'], real_data_population['total'], alpha = 0.5, color = 'orangered')
# ax.bar(real_data_population['age_group'], people_age_classes, alpha = 0.5, color = 'royalblue')

xticks = ['$04$','$59$','$1014$','$1519$','$2024$','$2529$','$3034$','$3539$','$4044$',
          '$4549$','$5054$','$5559$','$6064$','$6569$','$70+$']
ax.set_xticks(ticks = x, labels = xticks)
ax.set_title(r'Age structure comparison, 2022', fontsize=24, pad = 10)
ax.set_xlabel(r'Age group', fontsize=20)
ax.set_ylabel(r'Number of people', fontsize=20)
ax.legend(loc='upper left', ncols=2)
# xticks = real_data_population['age_group'].to_list()
ax.set_ylim([0, 1.1e6]) 
ax.grid(True)


fig.tight_layout() 
fig.savefig('../output/fig3.pdf', dpi=800) 
fig.savefig('../output/fig3.png', dpi=800)
```

```python
data_females = pd.read_csv("../data/data_females_processed.csv", index_col = 0)
data_females = data_females[['age_group', '2010']]
data_females = data_females.rename(columns = {'2010': 'female'})

data_males = pd.read_csv("../data/data_males_processed.csv", index_col = 0)
data_males = data_males[['age_group', '2010']]
data_males = data_males.rename(columns = {'2010': 'male'})

population_spb_2010 = pd.DataFrame()
population_spb_2010['age_group_old'] = data_males['age_group']
population_spb_2010['total'] = data_males['male'] + data_females['female']
population_spb_2010['male'] = data_males['male']
population_spb_2010['female'] = data_females['female']
new_age_groups = ['04', '59', '1014', '1519', '2024', '2529', '3034', 
                  '3539', '4044', '4549', '5054',  '5559', '6064', '6569', '70+']
total_num = population_spb_2010.iloc[14:]['total'].sum()
male_num = population_spb_2010.iloc[14:]['male'].sum()
female_num = population_spb_2010.iloc[14:]['female'].sum()
last_row = ['70+', total_num, male_num, female_num]
population_spb_2010.drop(population_spb_2010.tail(4).index,
        inplace = True)
population_spb_2010.loc[len(population_spb_2010)] = last_row
population_spb_2010.insert(0, 'age_group', new_age_groups)
population_spb_2010.drop(columns='age_group_old')
population_spb_2010.to_csv('population_spb_2010', sep='\t')
```
